```python
# drop the 'Active' and 'FN' columns 
customersData = customersData.drop(['Active', 'FN'], axis = 1)
# drop the rows with NaN values in the 'club_member_status' column
customersData = customersData.dropna(subset=['club_member_status'])

customersData.head()
```

```python
# drop the rest of rows with null values in them
customersData = customersData.dropna(subset=['fashion_news_frequency'])
customersData = customersData.dropna(subset=['age'])

# validate that there are no more null values in the data 
customersData.isnull().sum()
```

```python
# eliminate rows of data of people who are not between ages 20 and 40 years old
customersData = customersData[(customersData['age'] > 19) & (customersData['age'] < 41)]
# display the amount of customers for each age from 20 to 40 years old
customersData['age'].value_counts()
```

```python
transactionsData = pd.read_csv('/Users/aaronmcdonald/hm-fashion-recommender/data/raw/transactions_train.csv')
transactionsData.head(15)
```

```python
# Count the number of transactions for each customer
frequencyOf_customers = transactionsData['customer_id'].value_counts()

# Set a minimum threshold for the number of transactions a customer must have
min_transactions = 5

# Filter customer IDs that meet or exceed the minimum transaction threshold
customerIds = frequencyOf_customers[frequencyOf_customers >= min_transactions].index
```

```python
# Filter the transactions to include only those made by customers with at least 5 transactions
filtered_transactions = transactionsData[transactionsData['customer_id'].isin(customerIds)]

# Count the number of unique customers remaining after filtering
customers = filtered_transactions['customer_id'].nunique()

# Print the result
print('There are', customers, 'customers with a recorded transaction history of at least 5 transactions.')
```

```python
# filtered the data for only transactions associated with customers who are aged 20-40
filtered_transactions = filtered_transactions[filtered_transactions['customer_id'].isin(customersData['customer_id'])]

obs = filtered_transactions['customer_id'].nunique()
print('There are', obs, 'H&M customers aged 20-40 who have a recorded transaction history at least 5 total transactions')
print('In our target population there are', filtered_transactions['customer_id'].count(), 'different transactions among H&M customers within our desired demographic for the study.')
```

```python
final_transactions = filtered_transactions.sort_values('customer_id')
print(final_transactions.shape)
final_transactions.info()
```

```python
final_customersData = customersData[customersData['customer_id'].isin(final_transactions['customer_id'])]
print(final_customersData.shape)
customersData.info()
```

```python
articlesData = pd.read_csv('/Users/aaronmcdonald/hm-fashion-recommender/data/raw/articles.csv')

final_articlesData = articlesData[articlesData['article_id'].isin(final_transactions['article_id'])]
# Make sure all columns are shown in .info()
pd.set_option('display.max_columns', None)
final_articlesData.info()
```

```python
# Merge the dataframes
df = transactionsData.merge(final_customersData, on='customer_id', how='inner')
df = df.merge(articlesData, on='article_id', how='left')
```

```python
interaction_matrix = df.groupby(['customer_id', 'article_id']).size().unstack(fill_value=0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Normalize (optional)
interaction_matrix_normalized = interaction_matrix.div(interaction_matrix.sum(axis=1), axis=0)

# Compute similarity
item_similarity = cosine_similarity(interaction_matrix_normalized.T)
```

```python
# Pick a user
user_id = interaction_matrix.index[0]
user_vector = interaction_matrix.loc[user_id].values  # as numpy array

# Compute item scores using similarity matrix
scores = item_similarity.dot(user_vector)

# Map indices back to item IDs
item_ids = interaction_matrix.columns  # list of article_ids
user_vector_series = pd.Series(user_vector, index=item_ids)

# Remove already purchased items
already_purchased = user_vector_series[user_vector_series > 0].index
scores = pd.Series(scores, index=item_ids)
scores.loc[already_purchased] = -1  # exclude items already interacted with

# Get top 10 recommendations
recommended_items = scores.sort_values(ascending=False).head(10)

# Display article_ids
print("Top 10 Recommended Items for User", user_id)
print(recommended_items.index.tolist())
```
